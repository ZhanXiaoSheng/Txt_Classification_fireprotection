```python
def stop_words_list(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r').readlines()]
    return stopwords
```

```python
def txt_process(filename,data):
    import jieba
    file=open(filename,'a')
    stopwords=stop_words_list('stop_words.txt')
    for i in range(len(data)):
        s=data[i]
        s=re.sub(r"[A-Za-z0-9\：\;\１\２\０\４\５\６\７\８\３\/\\\+\９\:\(\)\（\）\-\·\—\，\；\.\、\,\～\【\】\。\n\t\“ \”]", "",s)
        s=jieba.cut(s.strip()) #移除字符串首尾的空格或换行
        outstr=''
        for word in s:
            if word not in stopwords:
                outstr+=word
                outstr+=' '
#             elif word==null:
#                 outstr+='设备报警'
        file.write(outstr+'\n')
    file.close()    
```

```python
def calSimilarity(x,y):  #计算两个向量得相似度
    if len(x)!=len(y):
        raise "维数不一样"
    c=0
    for i in  range(len(x)):
        c+=pow((x[i]-y[i]),2)
    return math.sqrt(c)
```

```python
def getWinner(x,layers):#找到layers里面与x最相似得节点
    min_value=1000000#存储最小距离
    min_index=-1#存储竞争层节点index
    for i in range(len(layers)):
        v=calSimilarity(x,layers[i])
        if v<min_value:
            min_value=v
            min_index=i
    return min_index
```

```python
#调整权值
def adjustWeight(w,x,alpha):  #w为要调整得权值向量；x为输入向量，alpha为学习率
    if len(w)!=len(x):
        raise "w,x 维度应相等"
    w_new=[]
    for i in range(len(w)):
        w_new.append(w[i]+alpha*(x[i]-w[i]))
    return w_new
    
```

```python
class SOM(object):
    def __init__(self, X, output, iteration, batch_size):

       # :param X:  形状是N*D， 输入样本有N个,每个D维
       # :param output: (n,m)一个元组，为输出层的形状是一个n*m的二维矩阵
       # :param iteration:迭代次数
       # :param batch_size:每次迭代时的样本数量
       # 初始化一个权值矩阵，形状为N*(n*m)，即有n*m权值向量，每个N维

        self.X = X
        self.output = output
        self.iteration = iteration
        self.batch_size = batch_size
        self.W = np.random.rand(X.shape[1], output[0]*output[1])
        print(self.W.shape)

    def GetN(self, t):
        #:param t:时间t, 这里用迭代次数来表示时间
        # return: 返回一个整数，表示拓扑距离，时间越大，拓扑邻域越小
        #
        a = min(self.output)
        return int(a-float(a)*t/self.iteration)

    def Geteta(self, t, n):
        # param t: 时间t, 这里用迭代次数来表示时间
        # param n: 拓扑距离
        # return: 返回学习率，
        return np.power(np.e, -n)/(t+2)

    def update_W(self, X, t, winner):
        N = self.GetN(t)
        for x, i in enumerate(winner):
            to_update = self.getneighbor(i[0], N)
            for j in range(N+1):
                e = self.Geteta(t, j)
                for w in to_update[j]:
                    self.W[:, W] = np.add(
                        self.W[:, W], e*(X[x, :]-self.W[:, w]))

    def getneighbor(self, index, N):
        # param index:获胜神经元的下标
        # param N: 邻域半径
        # return ans: 返回一个集合列表，分别是不同邻域半径内需要更新的神经元坐标
        a, b = self.output
        length = a*b

        def distance(index1, index2):
            i1_a, i1_b = index1//a, index1 % b
            i2_a, i2_b = index2//a, index2 % b
            return np.abs(i1_a-i2_a), np.abs(i1_b-i2_b)

        ans = [set() for i in range(N+1)]
        for i in range(length):
            dist_a, dist_b = distance(i, index)
            if dist_a <= N and dist_b <= N:
                ans[max(dist_a, dist_b)].add[i]
        return ans

    def train(self):
        # train_Y:训练样本与形状为batch_size*(n*m)
        # winner:一个一维向量，batch_size个获胜神经元的下标
        #:return:返回值是调整后的W
        #
        count = 0
        while self.iteration > count:
            train_X = self.X[np.random.choice(
                self.X.shape[0], self.batch_size)]
            normal_W(self.W)
            normal_X(train_X)
            train_Y = train_X.dot(self.W)
            winner = np.argmax(train_Y, axis=1).tolist()
            self.update_W(train_X, count.winner)
            count += 1
        return self.W

    def train_result(self):
        normal_X(self.X)
        train_Y = self.X.dot(self.W)
        winner = np.argmax(train_Y, axis=1).tolist()
        print(winner)
        return winner
```

```python
def normal_X(X):
    # X:二维矩阵，N*D，N个D维的数据
    # return: 将X归一化的结果
    N, D = X.shape
    for i in range(N):
        temp = np.sum(np.multiply(X[i], X[i]))
        X[i] /= np.sqrt(temp)
    return X
```

```python
def normal_W(W):
        # param W:二维矩阵，D*(n*m)，D个n*m维的数据
        # return: 将W归一化的结果
        #
    for i in range(W.shape[1]):
        temp = np.sum(np.multiply(W[:, i], W[:, i]))
        W[:, i] /= np.sqrt(temp)
    return W
```

```python
#初始化输入层与竞争层神经元的连接权值矩阵 
def initCompetition(n , m , d): 
    #随机产生0-1之间的数作为权值 
    array = random.random(size=n * m *d) 
    com_weight = array.reshape(n,m,d) 
    return com_weight
```

```python
#计算向量的二范数
def cal2NF(X):
    res = 0
    for x in X:
        res += x*x
    return res ** 0.5
```

```python
#对数据集进行归一化处理
def normalize(dataSet): 
    old_dataSet = copy(dataSet) 
    for data in dataSet: 
        two_NF = cal2NF(data) 
        for i in range(len(data)):
            data[i] = data[i] / two_NF 
    return dataSet , old_dataSet
```

```python
#对权值矩阵进行归一化处理 
def normalize_weight(com_weight): 
    for x in com_weight: 
        for data in x: 
            two_NF = cal2NF(data) 
            for i in range(len(data)): 
                data[i] = data[i] / two_NF 
    return com_weight
```

```python
#得到获胜神经元的索引值 
def getWinner(data , com_weight): 
    max_sim = 0 
    n,m,d = shape(com_weight) 
    mark_n = 0 
    mark_m = 0 
    for i in range(n): 
        for j in range(m):
            if sum(data * com_weight[i,j]) > max_sim: 
                max_sim = sum(data * com_weight[i,j]) 
                mark_n = i 
                mark_m = j 
    return mark_n , mark_m
```

```python
#得到神经元的N邻域 
def getNeibor(n , m , N_neibor , com_weight): 
    res = [] 
    nn,mm , _ = shape(com_weight)
    for i in range(nn): 
        for j in range(mm): 
            N = int(((i-n)**2+(j-m)**2)**0.5) 
            if N<=N_neibor: 
                res.append((i,j,N)) 
    return res
```
